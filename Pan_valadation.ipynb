```python
#Create a function for recent adjecent alphebet
def adjecent_recent_alphabet(pan):
    return any(pan[i] == pan[i+1] for i in range (len(pan)-1))
```

```python
#Create a fumction for checking secuencial alphebet 
def check_sequencial(pan):
    return all(ord(pan[i+1])-ord(pan[i])==1 for i in range(len(pan)-1))
```

```python
###MAIN FUNCTION###
```

```python
def is_valid_pan(pan):
    if len(pan)!=10:
        return False
    if not re.match(r'^[A-Z]{5}[0-9]{4}[A-Z]$',pan):
        return False
    if adjecent_recent_alphabet(pan):
        return False
    if  check_sequencial(pan):
        return False

    return True    
```

```python
df['Status'] = df['Pan_Numbers'].apply(lambda x:'Valid' if is_valid_pan(x) else 'Invalid')
df.head(10)
```

```python
df
```

```python
Valid_cnt = (df['Status']=='Valid').sum()
Valid_cnt
```

```python
Invalid_cnt = (df['Status']=='Invalid').sum()
Invalid_cnt
```

```python
Missing_cnt = 10000-(Valid_cnt+Invalid_cnt)
Missing_cnt
```

```python
df_summery =  pd.DataFrame({
    "TOTAL PROCESSED RECORD": 10000,
    "TOTAL VALID COUNT":[Valid_cnt],
    "TOTAL INVALID COUNT":[Invalid_cnt],
    "TOTAL MISSING COUNT":[Missing_cnt]

})
df_summery.head()
```

```python
with pd.ExcelWriter("PAN VALIDATION RESULT.xlsx") as writer :
    df.to_excel(writer,sheet_name='Pan Validations',index=False)
    df_summery.to_excel(writer,sheet_name='SUMMERY',index=False)
```
